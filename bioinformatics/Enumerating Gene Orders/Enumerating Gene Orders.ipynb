## Learning from other Solutions

```python
def permutate(orig_arr, temp_arr):
    if len(temp_arr) == len(orig_arr):
        print(" ".join(map(str, temp_arr)))
        temp_arr = []
    else:
        for item in orig_arr:
            if item not in temp_arr:
                temp_arr.append(item)
                permutate(orig_arr, temp_arr)
                temp_arr.remove(item)

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

num = 2
print(factorial(num))
permutate(list(range(1, num + 1)), [])
```
